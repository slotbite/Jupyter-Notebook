## Como se distribuye la informacin ?

Cada fichero se particiona en bloques 1,2,3,4 , y reparten en los dsitintos nodos.A su ves cada bloque  es replicado segun el factor de replicacin en genera el 3 , esto significa que hay 3 copias de un blouqe repartidas en los nodos , esto reprecenta la tolenacia a fallos de hadoop.



## Pero eso no es todo como , Como se gestionan la peticiones de escritura dentro del HDFS ?

### NameNode  
El namenode contiene la metadata, pero que es la metadata ? , son los datos relacionados con cada bloque como su tamao peso nombre , en que nodo se almacena y otras caracteristicas que lo identifican.
Al namenode llegan las peticiones de escritura y procesamiento de los datos. 
 



 

## Como de leen los datos ?

Desde el cliente las peticiones llegan al Namenode para ir a buscar los datos, es una especia de indice de datos por lo que debe ser lo mas rapido posible.




# Comandos HDSF

- ls : Lista el condenido de los directorios
- count :Cuneta el numero de directorios , ficheros , bytes en un dir.
- chgrp , chown y chmod : Cambia los permisos de ficheros y directorios.
- cat y text : Muestra el contenido de ficheros.
- tail :Muestra el ultimo KB del contenido del fichero.
- get y copyTolocal : Copia un fichero o directorio DESDE HDFS al sistema local.
- put y copyFromlocal : Copia un fichero a directorio local HACIA HDFS.
- getMerge : Obtiene de dos ficheros en un unico fichero y lo almacena en el sistema de ficheros local.
- mv :mueve ficheros y directorios en HDFS
- cp : Copia ficheros y directorios em HDFS.
- mkdir : Crea directorios nuevos en HDFS.
- rm : Borra un fichero en HDFS (a papelera).
- rm -R : Borra un directorio recursivamente en HDFS (a papelera)
